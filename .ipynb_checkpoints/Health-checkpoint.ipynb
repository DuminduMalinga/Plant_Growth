```python
import cv2
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt  # ✅ For displaying images in Jupyter Notebook

# Define paths
image_folder = "Plant_Images/Day7_Evening/Health"

# ✅ Check if the image folder exists before proceeding
if not os.path.exists(image_folder):
    print(f"⚠️ Error: Folder '{image_folder}' not found. Please check the path.")
    exit()

# List all image files in the folder
image_files = [f for f in os.listdir(image_folder) if f.lower().endswith((".jpg", ".png", ".jpeg"))]

# Function to segment plant and analyze color changes
def analyze_color_with_mask(image_path):
    # Load the image
    image = cv2.imread(image_path)
    
    # ✅ Check if the image is loaded correctly
    if image is None:
        print(f"⚠️ Error: Could not load image {image_path}. Skipping...")
        return None, None, None

    # Convert to HSV color space
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Convert image to grayscale for segmentation
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur to remove noise
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Apply Otsu's thresholding to create a binary mask
    _, binary_mask = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # Apply mask to the HSV image (only select plant pixels)
    masked_hsv = cv2.bitwise_and(hsv, hsv, mask=binary_mask)

    # Extract the non-zero pixels (only plant region)
    hue_values = masked_hsv[:, :, 0][binary_mask > 0]
    saturation_values = masked_hsv[:, :, 1][binary_mask > 0]
    value_values = masked_hsv[:, :, 2][binary_mask > 0]

    # Compute mean values for Hue, Saturation, and Value (only for plant pixels)
    mean_hue = np.mean(hue_values) if len(hue_values) > 0 else 0
    mean_saturation = np.mean(saturation_values) if len(saturation_values) > 0 else 0
    mean_value = np.mean(value_values) if len(value_values) > 0 else 0

    # ✅ Convert image to RGB for displaying in Jupyter Notebook
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # ✅ Display the original and masked images
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    axes[0].imshow(image_rgb)
    axes[0].set_title(f"Original Image: {os.path.basename(image_path)}")
    axes[0].axis("off")

    axes[1].imshow(binary_mask, cmap="gray")
    axes[1].set_title("Segmented Mask")
    axes[1].axis("off")

    plt.show()

    return mean_hue, mean_saturation, mean_value

# Analyze all images and store results
color_analysis_results = []

for image_file in image_files:
    image_path = os.path.join(image_folder, image_file)
    
    # Analyze color changes using a mask
    mean_hue, mean_saturation, mean_value = analyze_color_with_mask(image_path)
    
    if mean_hue is not None:  # Ensure valid data before storing
        color_analysis_results.append({
            "Image": image_file,
            "Mean Hue": mean_hue,
            "Mean Saturation": mean_saturation,
            "Mean Brightness (Value)": mean_value
        })

# Convert results into a DataFrame
df_results = pd.DataFrame(color_analysis_results)

# Display the final report
print("\n--- Color Analysis Report (Using Mask) ---\n")
print(df_results)

# ✅ Define the correct path to save the report inside the image folder
excel_path = os.path.join(image_folder, "Color_Analysis_Report_Masked_Day07Evening.xlsx")

# ✅ Save the report to the specified folder
df_results.to_excel(excel_path, index=False, engine="openpyxl")

print(f"\nReport saved as '{excel_path}'")
```
